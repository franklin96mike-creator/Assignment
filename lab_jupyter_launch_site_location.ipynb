```python
# TASK 3: Calculate the distances between a launch site to its proximities

import pandas as pd
import numpy as np

# Function to calculate distance between two lat/long points (Haversine formula)
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    
    a = np.sin(delta_phi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    
    return R * c

# Calculate distance from each launch site to the reference coordinate
spacex_df['Distance_km'] = spacex_df.apply(
    lambda row: haversine_distance(row['Lat'], row['Long'], nasa_coordinate[0], nasa_coordinate[1]),
    axis=1
)

# Display first 5 distances
spacex_df[['Launch Site', 'Lat', 'Long', 'Distance_km']].head()
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon
import numpy as np

# Function to calculate Haversine distance
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    
    a = np.sin(delta_phi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    
    return R * c

# Example feature coordinates (replace with actual coordinates from MousePosition)
city_coord = [28.3922, -80.6077]       # Closest city
railway_coord = [28.5000, -80.6000]    # Closest railway
highway_coord = [28.5800, -80.5500]    # Closest highway

# Create map
site_map = folium.Map(location=[28.5, -80.57], zoom_start=6)
MousePosition().add_to(site_map)

# Marker cluster for launch sites
marker_cluster = MarkerCluster().add_to(site_map)

# Plot all launch sites
for idx, row in spacex_df.iterrows():
    launch_lat = row['Lat']
    launch_lon = row['Long']
    
    # Add launch site marker
    folium.Marker(
        location=[launch_lat, launch_lon],
        popup=f"Launch Site: {row['Launch Site']}",
        icon=folium.Icon(color='blue')
    ).add_to(marker_cluster)
    
    # Draw line and distance to closest city
    dist_city = calculate_distance(launch_lat, launch_lon, *city_coord)
    folium.PolyLine(
        locations=[[launch_lat, launch_lon], city_coord],
        color='blue', weight=1, opacity=0.7
    ).add_to(site_map)
    folium.Marker(
        location=[launch_lat, launch_lon],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 10pt; color:blue;"><b>%s KM</b></div>' % "{:.2f}".format(dist_city)
        )
    ).add_to(site_map)
    
    # Draw line and distance to closest railway
    dist_railway = calculate_distance(launch_lat, launch_lon, *railway_coord)
    folium.PolyLine(
        locations=[[launch_lat, launch_lon], railway_coord],
        color='purple', weight=1, opacity=0.7
    ).add_to(site_map)
    
    # Draw line and distance to closest highway
    dist_highway = calculate_distance(launch_lat, launch_lon, *highway_coord)
    folium.PolyLine(
        locations=[[launch_lat, launch_lon], highway_coord],
        color='red', weight=1, opacity=0.7
    ).add_to(site_map)

# Add markers for the features
folium.Marker(city_coord, popup="Closest City", icon=folium.Icon(color='blue')).add_to(site_map)
folium.Marker(railway_coord, popup="Closest Railway", icon=folium.Icon(color='purple')).add_to(site_map)
folium.Marker(highway_coord, popup="Closest Highway", icon=folium.Icon(color='red')).add_to(site_map)

# Display map
site_map
```
